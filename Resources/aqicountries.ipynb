```python
import pandas as pd
import requests
```

```python
og_df = pd.read_csv("./who_ambient_air_quality.csv")
og_df
```

```python
columns_selection = ["country_name","pm25_concentration","year"]
clean_df = og_df[columns_selection].dropna(subset=["pm25_concentration"])
gb = clean_df.groupby(["country_name","year"]).mean().reset_index()
gb
```

```python
countrygroup = og_df.groupby(['country_name'])
lat=countrygroup['latitude'].first()
long=countrygroup['longitude'].first()
coordinates_df = pd.DataFrame({'latitude':lat, "longitude":long})
coordinates_df.head()
```

```python
merge_df = pd.merge(gb,coordinates_df,on='country_name')
merge_df.head()
```
